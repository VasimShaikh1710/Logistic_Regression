```python
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\admin\\Desktop\\Assignment_DataScience\\bank-full.csv",sep=';')
df
```

```python
df.info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
a = ['job','marital','education','contact','poutcome','month']
```

```python
ohe = OneHotEncoder(drop = 'first', sparse= False)

for i in a:
    df[i] = ohe.fit_transform(df[[i]])
```

```python
df
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
b = ['default', 'housing', 'loan','y']
```

```python
le = LabelEncoder()
for i in b:
    df[i] = le.fit_transform(df[i])
```

```python
df
```

```python
df.info()
```

```python
x=pd.concat([df.iloc[:,0:10],df.iloc[:,11:]],axis=1)
x
```

```python
y = df.iloc[:,10]
y
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier=LogisticRegression()
classifier.fit(x,y)
```

```python
y_pred=classifier.predict(x)
y_pred
```

```python
y_pred_df=pd.DataFrame({'actual_y':y,'y_pred_prob':y_pred})
y_pred_df
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, roc_curve, roc_auc_score
```

```python
# Confusion Matrix for the model accuracy
confusion_matrix = confusion_matrix(y,y_pred)
confusion_matrix
```

```python
AS = accuracy_score(y,y_pred)
AS.round(2)
```

```python
classifier.predict_proba(x)[:,1] 
```

```python
import matplotlib.pyplot as plt 
```

```python
# ROC Curve plotting and finding AUC value
fpr,tpr,thresholds=roc_curve(y,classifier.predict_proba(x)[:,1])
plt.plot(fpr,tpr,color='red')
auc=roc_auc_score(y,y_pred)

plt.plot(fpr,tpr,color='red',label='logit model(area  = %0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.show()

print('auc accuracy:',auc)
```
